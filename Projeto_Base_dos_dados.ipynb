```python
#Importando as bibliotecas utilizadas:
import basedosdados as bd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Carregando a tabela do data lake:
df = bd.read_table(dataset_id='br_ibge_censo_agropecuario',
table_id='municipio',
billing_project_id='projeto11-360320')
```

```python
#verificando as primeiras linhas:
df.head()
```

```python
#Selecionando as colunas que serão utilizadas:
proj = df[['ano', 'sigla_uf', 'producao_cafe']]
display(proj)
```

```python
#Pegando apenas uma parte da tabela para fazer o grafico, pois havia muitas linhas.
base_final = proj[15000:20738]
display(base_final)
```

```python
#Para fazer o grafico, aumentei o espaçamento, coloquei as bases que compoe  grafico e escolhi a cor.
# Depois coloquei titulo para o grafico e para o valor de X e Y.
plt.figure(figsize=(20,6))
plt.plot(base_final['sigla_uf'], base_final['producao_cafe'],color = 'green')
plt.ylabel('Produção do Café')
plt.xlabel('Estados')
plt.title('Produção do Café por Estado', fontsize = 14)
plt.show()
```
